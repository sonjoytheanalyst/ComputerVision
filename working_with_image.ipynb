## Affine Tranformation

```python
import tensorflow as tf
transformation=tf.keras.preprocessing.image.apply_affine_transform(img, theta=45)

plt.imshow(transformation)
plt.show()
```

```python

transformation=tf.keras.preprocessing.image.apply_affine_transform(img, zx=2,zy=2)

plt.imshow(transformation)
plt.show()
```

```python
 #  tx=Width shift.
  #  ty=Heigh shift.
transformation=tf.keras.preprocessing.image.apply_affine_transform(img, tx=250,ty=5)

plt.imshow(transformation)
plt.show()
```
